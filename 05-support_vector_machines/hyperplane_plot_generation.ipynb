```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
import numpy as np

data, target = make_blobs(n_samples=300, n_features=3, centers=2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

colors = ['#FE4365' if x==0 else '#83AF9B' for x in target]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=data[:,0], ys=data[:,1], zs=data[:,2],
           c=colors, edgecolors='b', linewidth=0.2)

# hide axis ticks
"""
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.zaxis.set_ticks([])
"""

plt.show()
```

```python
clf = SVC(kernel='linear')
clf.fit(data, target)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

z = lambda x,y: (-clf.intercept_[0]-clf.coef_[0][0]*x-clf.coef_[0][1]) / clf.coef_[0][2]

tmp = np.linspace(-2,10,51)
x,y = np.meshgrid(tmp,tmp)
ax.plot_surface(x, y, z(x,y), alpha=0.8, color='#F9CDAD')
ax.scatter(xs=data[:,0], ys=data[:,1], zs=data[:,2], 
           c=colors, edgecolors='b', linewidth=0.2)

# hide axis ticks
"""
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.zaxis.set_ticks([])
"""

plt.show()
```
